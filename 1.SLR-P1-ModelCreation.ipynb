```python
#Import pandas python library, this library is used to access and manipulate datasets like a spreadsheet, csv, sql #and many more.
import pandas as pd
```

```python
#Read the csv file using pandas')
df = pd.read_csv('Salary_Data.csv')
```

```python
df #Display the dataset
```

```python
#split input as 'x' and output as 'y'
x_value = df[['YearsExperience']].values
y_value = df['Salary'].values
```

```python
# Install matplotlib package using pip
%pip install matplotlib

# Import matplotlib.pyplot for plotting
import matplotlib.pyplot as plt
```

```python
# Create a scatter plot of Years of Experience vs Salary
plt.scatter(x_value, y_value, color='blue', marker='o')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs Salary')
plt.show()
```

```python
#Install scikit-learn package using pip
%pip install scikit-learn
```

```python
#split the dataset into training and testing sets
from sklearn.model_selection import train_test_split as tts
```

```python
x_train,x_test,y_train,y_test=tts(x_value, y_value, test_size=0.3, random_state=42) #Split the dataset into training and testing sets
print("Training set size:", len(x_train))
print("Training input values", x_train) 
print("Training output values", y_train)
print("Testing set size:", len(x_test)) 
print("Testing input values", x_test)
print("Testing output values", y_test)
```

```python
#create a linear regression model
from sklearn.linear_model import LinearRegression as lr
```

```python
model = lr() #Create a linear regression model
model.fit(x_train, y_train) #Fit the model with training data
```

```python
#view the coefficients and intercept of the model
coef = model.coef_
print("Coefficient:", coef)
intercept = model.intercept_
print("Intercept:", intercept)
```

```python
#test the model with the testing data
y_pred = model.predict(x_test) #Predict the salary using the testing data
#view the predicted values
print("Predicted Salaries:", y_pred)
#view the actual values
print("Actual Salaries:", y_test)
#plot the regression line
plt.scatter(x_value, y_value, color='blue', marker='o', label='Actual Data')
plt.plot(x_test, y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Years of Experience') 
plt.ylabel('Salary')
plt.title('Regression Line for Salary Prediction')
```

```python
from sklearn.metrics import r2_score # Import r2_score to calculate the R-squared value
# Calculate the R-squared value
r2 = r2_score(y_test, y_pred)
print("R-squared value:", r2)
```

```python
#Import the pickle library to save the model
import pickle as pkl
# Save the trained model to a file
pkl.dump(model, open('salary_model.sav', 'wb')) #Save the model to a file named 'salary_model.sav'
```

```python
# Load the model from the file
loaded_model = pkl.load(open('salary_model.sav', 'rb'))
# Use the loaded model to make predictions
new_data = 5  # Example input for prediction (5 years of experience)
predicted_salary = loaded_model.predict([[new_data]])
print("Predicted Salary for ", new_data," years of experience:", predicted_salary[0]) 
```
